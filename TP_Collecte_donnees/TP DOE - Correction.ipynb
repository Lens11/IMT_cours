## Exercise 1

```python
import pandas as pd
tab = pd.DataFrame({'A': [-1, -1, 1, 1],
                    'B': [-1, 1, -1, 1],
                    'Y': [6, 7.4, 10.8, 11.2]})
tab
```

(b)

```python
tab['AB'] = tab['A'] * tab['B']
tab = tab[['A', 'B', 'AB', 'Y']]
tab
```

(c)

```python
effect_a = 2 * (tab.A * tab.Y).mean()
effect_b = 2 * (tab.B * tab.Y).mean()
effect_ab = 2 * (tab.AB * tab.Y).mean()
gen_effect = tab.Y.mean()
print(effect_a)
print(effect_b)
print(effect_ab)
print(gen_effect)
```

(d)

```python
def pred_func(a, b):
  return gen_effect + (effect_a*a + effect_b*b + effect_ab*a*b)/2
```

(e)

```python
tab['pred_fact'] = pred_func(tab.A, tab.B)
tab
```

(f)

```python
from sklearn.linear_model import LinearRegression
X = tab[['A', 'B']]
y = tab.Y
reg = LinearRegression()
reg.fit(X, y)
tab['preds_lin'] = reg.predict(X)
tab
```

(g)

```python
import numpy as np
rmse_fact = np.sqrt(np.mean((tab.Y-tab.pred_fact)**2))
rmse_lin = np.sqrt(np.mean((tab.Y-tab.preds_lin)**2))
print('RMSE(factorial model) =', rmse_fact)
print('RMSE(linear model) =', rmse_lin)
```

(h)

```python
def pred_func2(a, b):
  return gen_effect + (effect_a*a + effect_b*b)/2
```

```python
tab['pred_fact2'] = pred_func2(tab.A, tab.B)
tab
```

```python
rmse_fact2 = np.sqrt(np.mean((tab.Y-tab.pred_fact2)**2))
print(rmse_fact2)
```

```python
def coef_cst(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
             ABC_effect, general_effect):
    return general_effect
- (A_effect/2)*(5/3) - (B_effect/2)*(45/35) - (C_effect/2)*1.5
+(AB_effect/2)*(5/3)*(45/35)+(AC_effect/2)*(5/3)*1.5+(BC_effect/2)*(45/35)*1.5
-(ABC_effect/2)*(5/3)*(45/35)*1.5

def coef_A(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
           ABC_effect, general_effect):
    return (A_effect/2)*(1/3)
-(AB_effect/2)*(1/3)*(45/35)-(AC_effect/2)*(1/3)*1.5
+(ABC_effect/2)*(1/3)*(45/35)*1.5

def coef_B(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
           ABC_effect, general_effect):
    return (B_effect/2)*(1/35)
-(AB_effect/2)*(5/3)*(1/35)-(BC_effect/2)*(1/35)*1.5
+(ABC_effect/2)*(5/3)*(1/35)*1.5

def coef_C(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
           ABC_effect, general_effect):
    return (C_effect/2)*1.5
-(AC_effect/2)*(5/3)-(BC_effect/2)*(45/35)
+(ABC_effect/2)*(5/3)*(45/35)

def coef_AB(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
            ABC_effect, general_effect):
    return (AB_effect/2)*(1/3)*(1/35)-(ABC_effect/2)*(1/3)*(1/35)*1.5

def coef_AC(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
            ABC_effect, general_effect):
    return (AC_effect/2)*(1/3)-(ABC_effect/2)*(1/3)*(45/35)

def coef_BC(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
            ABC_effect, general_effect):
    return (BC_effect/2)*(1/35)-(ABC_effect/2)*(5/3)*(1/35)

def coef_ABC(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect,
             ABC_effect, general_effect):
    return (ABC_effect/2)*(1/3)*(1/35)
```

```python
def completeDOE(a,b,c):
    cst = coef_cst(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefA = coef_A(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefB = coef_B(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefC = coef_C(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefAB = coef_AB(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefAC = coef_AC(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefBC = coef_BC(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    coefABC = coef_ABC(A_effect, B_effect, C_effect, AB_effect, AC_effect, BC_effect, ABC_effect, general_effect)
    return cst + coefA*a + coefB*b + coefC*c + coefAB*a*b + coefAC*a*c + coefBC*b*c + coefABC*a*b*c
```

```python
def latinDOE(a,b,c):
    cst = coef_cst(A_latin_effect, B_latin_effect, C_latin_effect,
                   AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefA = coef_A(A_latin_effect, B_latin_effect, C_latin_effect,
                   AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefB = coef_B(A_latin_effect, B_latin_effect, C_latin_effect,
                   AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefC = coef_C(A_latin_effect, B_latin_effect, C_latin_effect,
                   AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefAB = coef_AB(A_latin_effect, B_latin_effect, C_latin_effect,
                     AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefAC = coef_AC(A_latin_effect, B_latin_effect, C_latin_effect,
                     AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefBC = coef_BC(A_latin_effect, B_latin_effect, C_latin_effect,
                     AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    coefABC = coef_ABC(A_latin_effect, B_latin_effect, C_latin_effect,
                       AB_latin_effect, AC_latin_effect, BC_latin_effect, ABC_latin_effect, general_latin_effect)
    return cst + coefA*a + coefB*b + coefC*c + coefAB*a*b + coefAC*a*c + coefBC*b*c + coefABC*a*b*c
```

```python
def randomDOE(a,b,c):
    cst = coef_cst(A_rand_effect, B_rand_effect, C_rand_effect,
                   AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefA = coef_A(A_rand_effect, B_rand_effect, C_rand_effect,
                   AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefB = coef_B(A_rand_effect, B_rand_effect, C_rand_effect,
                   AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefC = coef_C(A_rand_effect, B_rand_effect, C_rand_effect,
                   AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefAB = coef_AB(A_rand_effect, B_rand_effect, C_rand_effect,
                     AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefAC = coef_AC(A_rand_effect, B_rand_effect, C_rand_effect,
                     AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefBC = coef_BC(A_rand_effect, B_rand_effect, C_rand_effect,
                     AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    coefABC = coef_ABC(A_rand_effect, B_rand_effect, C_rand_effect,
                       AB_rand_effect, AC_rand_effect, BC_rand_effect, ABC_rand_effect, general_rand_effect)
    return cst + coefA*a + coefB*b + coefC*c + coefAB*a*b + coefAC*a*c + coefBC*b*c + coefABC*a*b*c
```
