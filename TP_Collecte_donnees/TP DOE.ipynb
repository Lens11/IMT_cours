
Question (c)
Pour calculer le computation effect, on suit les formules du cours (p38), cela sert à savoir si on augmente de 1 A, de combien va augmenter Y... Pour A, il faut faire A = (yi x ai)x2/N
.mean() fait la moyenne donc il divise par 4 (c'est pour ca qu'on multiplie par 2 encore)



```python
effect_A = (tab.A * tab.Y).mean()*2
effect_B = (tab.B * tab.Y).mean()*2
effect_AB = (tab.AB * tab.Y).mean()*2
gen_effect = tab.Y.mean()


print(effect_A)
print(effect_B)
print(effect_AB)
print(gen_effect)
```

Question (d) On va maintenant retrouver les valeurs de Y avec une fonction de prediction des valeurs de Y (qu'on va appliquer ligne par ligne)

```python
 def fonction_prediction(a,b):
  return gen_effect + (effect_A*a + effect_B*b + effect_AB*a*b)/2
```

On va maintenant appliquer cette fonction aux valeurs pour essayer de retrouver Y

```python
tab['pred_fact'] = fonction_prediction(tab.A, tab.B)
tab
```

On remarque que le modèle est parfait et que c'est assez rare. On va maintenant faire un modèle de regression linéaire

```python
from sklearn.linear_model import LinearRegression
X=tab[['A','B']]
Y=tab.Y
reg = LinearRegression()
reg.fit(X,Y)
tab['preds_lin'] = reg.predict(X)
tab
```

On voit que les modèles sont quasiment similaires, mais le modèle factoriel est meilleur

```python
import numpy as np
rmse_fact = np.sqrt(np.mean((tab.Y-tab.)**2))
```
