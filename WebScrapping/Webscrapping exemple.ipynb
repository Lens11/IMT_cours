```python
import requests as r 
from bs4 import BeautifulSoup
import pandas
```

```python
page = r.get('https://web.archive.org/web/20121007172955'+\
    '/https://www.nga.gov/collection/anZ1.htm')
page.text
```

```python
page_root = BeautifulSoup(page.text)
page_root
```

```python
div = page_root.find("div", attrs={'class': 'BodyText'})

rows = div.find("table").find_all("tr")

records = []

for row in rows:
  cols = row.find_all('td')
  link = "https://web.archive.org" + cols[0].find('a')['href']
  name = cols[0].find('a').text
  birth_info = cols[1].text
  nationality = birth_info.split(',')[0]
  dates = birth_info.split(',')[1]
  birth_date = -1
  death_date = -1
  period = "N/A"
  skip = False
  if "born" in dates:
    birth_date = int(dates.strip().split(" ")[1])
  elif "-" in dates and "c." not in dates:
    birth_date = int(dates.strip().split("-")[0].strip())
    death_date = int(dates.strip().split("-")[1].strip())
  elif "-" in dates and "c." in dates:
    birth_date = int(dates.strip().split("-")[0].strip().split(" ")[1])
    death_date = int(dates.strip().split("-")[1].strip())
  elif "active" in dates and "century" in dates:
    period = dates.split("active")[-1]
  else:
    skip=True
    # print(dates)
  records.append({'name': name, 'nationality': nationality, 'birth_date': birth_date, 'death_date': death_date, 'period': period})

print(records)
```

```python
from pandas import DataFrame
df = DataFrame.from_records(records)
df.to_csv("artistes_y1.csv")
```
