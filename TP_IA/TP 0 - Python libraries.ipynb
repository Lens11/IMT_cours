````python
import numpy as np

n = 3
A = np.random.rand(n,n) # random n*n matrix
B = np.zeros((n,n)) # n*n matrix filled with 0
c = np.ones(n) # vector of size n filled with 1
d = c + 1 # add scalar to all vector value
e = np.arange(n) # [0, 1,  ..., n-1]

X = np.c_[ c , B, d ] # concatenate column vectors to build a matrix

X_reduced = X[:, [0, 4]] # select a subset of the matrix wpecifying columns of interest, ':' means all (no restriction)

Y = np.array([[1,0,0],[2,0,0],[3,0,0]]) # build a matrix specifying raws

Z = Y @ X_reduced # matrix multiplication

Z_reduced = Z[[0,2],:]

print(f"A:\n{A}")
print(f"B:\n{B}")
print(f"c:\n{c}")
print(f"d:\n{d}")
print(f"e:\n{e}")
print(f"X:\n{X}")
print(f"X_reduced:\n{X_reduced}")
print(f"Y:\n{Y}")
print(f"Z:\n{Z}")
print(f"Z_reduced:\n{Z_reduced}")
````

Below an example of how Numpy can be used for linear algebra.

````python
A = np.array([[1.0, 2.0], [3.0, 4.0]])
print(f"A:\n{A}")

A_t = A.transpose()
print(f"A_t:\n{A_t}")
print(f"A.T:\n{A.T}")

A_i = np.linalg.inv(A)
print(f"A inverse:\n{A_i}")

U = np.eye(2) # unit 2x2 matrix; "eye" represents "I"
print(f"U:\n{U}")

J = np.array([[0.0, -1.0], [1.0, 0.0]])
print(f"J:\n{J}")

JJ = J @ J        # matrix product
print(f"JJ:\n{JJ}")

tru =  np.trace(U)  # trace
print(f"tr(u):\n{tru}")

Y = np.array([[5.], [7.]])
Y_solve = np.linalg.solve(A, Y)
print(f"Y_solve:\n{Y_solve}")

eig_J = np.linalg.eig(J)
print(f"eig_J:\n{eig_J}")
````

````python
%matplotlib inline
x = np.linspace(1, 30, 100)

plt.plot(x, np.log(x), label='linear $x$')
plt.plot(x, np.log(x**2), label='quadratic $x^2$')
plt.plot(x, np.log(x**3), label='cubic $x^3$')
#plt.plot(x, np.exp(x), label='exponential $e^x$')

plt.xlabel('number of inputs')
plt.ylabel('$log f(x)$')

plt.title("Log Complexity for various functions")

plt.legend()

plt.show()
````

````python
# Adapted from https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py
%matplotlib inline

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1) # the first figure

plt.subplot(212) # the second subplot in the first figure
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.figure(2)                # figure 1 current; subplot(212) still current
plt.subplot(211) 
plt.title('Histogram of IQ')

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)


plt.show()
````

````python
#From https://matplotlib.org/gallery/images_contours_and_fields/image_demo.html
%matplotlib inline
import matplotlib.cm as cm

delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
im = ax.imshow(Z, interpolation='bilinear', cmap=cm.RdYlGn,
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())

plt.show()
````
